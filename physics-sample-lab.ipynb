## Abstract

The angular acceleration of an rigid body is dependent upon four factors:

- magnitude of the force
- direction of the force
- point at which the force is applied to the object
- rotational inertia of the rigid body.

This lab experiment enabled us to exampne the application of a force applied to a rigid body that is applied to a fixed axis. The magnitude of the initia was changed btween three values to observe how the magnitude of the force (weight) impacted the angular acceleration of the object.

After rotating the plates, $T_1$, or tension 1 applies torque to the plate by the step pulley. $I_M$ is the moment of inertia of the plate and $\alpha$, is the angular acceleration of the plate and $r$ is the radius of the disk.

$$\tau=T_1 r=I_M\alpha$$

At the rotary motion detector, different tensions exist on either side of the pulley. In the following equation, $I_p$ is the moment of inertia of the motion sensor, $\alpha_2$ is the angular acceleration of the motion sensor pulley, $T_1$ and $T_2$ are the tensions of the string and $R$ is the radius of the motion sensor.

$$\tau_1-\tau_2 = T_2 R - T_1 R = I_p \alpha_2$$

Where mass $m$ is suspended we can use Newton's second law where $m$ is the mass, $T_2$ is the tension in the string and $a2$ is the acceleration of the object downward.
